## Assign Data Parameters to Musical Aspects

```python
import numpy as np

convert_element_to_aspect = ConvertElementToAspect(loaded_data)

# convert valence [-100, 100]
valence_array = convert_element_to_aspect.convert_element_to_valence('precipitation', 20, 0, isInverted=True)
print(valence_array)
# convert valence [0, 100]
arousal_array = convert_element_to_aspect.convert_element_to_arousal('temperature', 20, 10)
print(arousal_array)

e = ValenceArousalToEmotion(valence_array, arousal_array)
emotion_array = e.convert_valencea_arousal_to_emotion()
```

### Generate Music Based Using Suno API

```python
style = 'Electronic Music'
upload_url_base = 'https://audio-eval-2025-05.web.app/input_melody/'
upload_filenames = ["melody_1_val100_aro100", "melody_2_val-100_aro10", "melody_3_val100_aro100", "melody_4_val66_aro83", "melody_5_val33_aro66", "melody_6_val0_aro50", "melody_7_val-33_aro34", "melody_8_val-66_aro17", "melody_9_val-100_aro10"]

generator = SunoMusicGenerator()
for idx, emotion_param in enumerate(emotion_array):
    upload_url = f"{upload_url_base}{upload_filenames[idx]}.wav"
    print(upload_url)
    filename = generator.run(emotion_param, style, upload_url, upload_filenames[idx])
```
