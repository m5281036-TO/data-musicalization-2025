```python
from FileLoader import FileLoader
from WeatherPatternMining import WeatherPatternMining
from DisplayHistogram import DisplayHistogram
from RandomDataPicker import RandomDataPicker
from ConvertElementToAspect import ConvertElementToAspect
from PromptExporter import PromptExporter
from CreateChordsAndMelody import CreateChordsAndMelody
from ValenceArousalToEmotion import ValenceArousalToEmotion
# from MusicGenFromMelody import MusicGenFromMelody
from CreateCrossfade import CreateCrossFade
from SunoMusicGenerator import SunoMusicGenerator
from CrossfadeAudioFiles import CrossfadeAudioFiles
```

## Loading Data

```python
min = WeatherPatternMining()
preprocessed_data = min.preprocess_and_pattern_miming()
```

## Assign Data Parameters to Musical Aspects

```python
import numpy as np
import pandas as pd

convert_element_to_aspect = ConvertElementToAspect(preprocessed_data)

# convert valence [-100, 100]
valence_array = convert_element_to_aspect.convert_element_to_valence('temperature', 1, -1)
print(valence_array)
# convert valence [0, 100]
arousal_array = convert_element_to_aspect.convert_element_to_arousal('support', 0.1, 0.001)
print(arousal_array)

e = ValenceArousalToEmotion(valence_array, arousal_array)
emotion_array = e.convert_valencea_arousal_to_emotion()

emotion_df = pd.DataFrame({'valence': valence_array, 'arousal': arousal_array, 'emotion': emotion_array, 'temperature': preprocessed_data['temperature'], 'support':preprocessed_data['support']})
print(emotion_df)

rand = RandomDataPicker(emotion_df)
rand_df = rand.get_random_data_samples(5, 'emotion')
```

## Generate Music from Prompt
### Generate Melody Files (MIDI and WAV Files)

```python
c = CreateChordsAndMelody()
c.create_midi_and_wav(rand_df['valence'], rand_df['arousal'])
```

### Generate Music Based Using Suno API

```python
style = 'Electronic Music'
upload_url_base = 'https://audio-eval-2025-05.web.app/input_melody/'
upload_filenames = ["melody_1_val-100_aro90", "melody_2_val20_aro100", "melody_3_val-10_aro0", "melody_4_val100_aro0", "melody_5_val0_aro0"]

# generate tasks
generator = SunoMusicGenerator()
task_ids = []
for idx, emotion_param in enumerate(rand_df['emotion']):
    upload_url = f"{upload_url_base}{upload_filenames[idx]}.mp3"
    task_ids.append(generator.generate_music(emotion_param, style, upload_url))

print(task_ids)

# download generated tracks
for idx, task_id in enumerate(task_ids):
    audio_url = generator.poll_suno_task(task_id)
    filename = generator.download_tracks(audio_url, f'{upload_filenames[idx]}')
    
print(f"{len(task_ids)} task(s) processing: {task_ids}")
```
